```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
import pandas as pd
```

```python
dados = pd.read_csv("/content/ConsumoCo2 (1).csv")
dados.head()
```

```python
x = dados [['MODELYEAR', 'MAKE', 'MODEL', 'VEHICLECLASS', 'ENGINESIZE', 'CYLINDERS', 'TRANSMISSION', 'FUELTYPE', 'FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY', 'FUELCONSUMPTION_COMB', 'FUELCONSUMPTION_COMB_MPG']]
x = pd.get_dummies(x)
x = x.astype(float)
x.head()
```

```python
y = dados[['CO2EMISSIONS']]
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,
y, test_size=0.5, random_state=0)
gnb = GaussianNB()
y_pred = gnb.fit(x_train, y_train).predict(x_test).reshape(-1, 1)
```

```python
print("Number of mislabeled points out of a total",
x_test.shape[0], (y_test != y_pred).sum())
```
